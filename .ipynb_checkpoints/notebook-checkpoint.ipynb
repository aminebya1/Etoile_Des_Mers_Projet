```python
# le nombre de valeurs manquantes par colonne
missing_values = SeaCareerSet.isna().sum()
missing_values
```

```python
# Imputer les valeurs manquantes de 'embarked' en utilisant le mode des valeurs de 'class'
mode_embarked_per_class = SeaCareerSet.groupby('class')['embarked'].apply(lambda x: x.mode().iloc[0])
SeaCareerSet['embarked'] = SeaCareerSet.apply(
    lambda row: mode_embarked_per_class[row['class']] if pd.isna(row['embarked']) else row['embarked'],
    axis=1
)
```

```python
# Remplacer les valeurs manquantes de 'age' par la moyenne
SeaCareerSet['age'].fillna(SeaCareerSet['age'].mean(), inplace=True)
```

```python
# Remplacer les valeurs manquantes de 'fare' par la médiane
SeaCareerSet['fare'].fillna(SeaCareerSet['fare'].median(), inplace=True)
```

```python
# Remplacer les valeurs manquantes de 'embarked' par le mode
SeaCareerSet['embarked'].fillna(SeaCareerSet['embarked'].mode()[0], inplace=True)
```

#  EX 9

```python
def extract_title(name):
    title_part = name.split(',')[1].strip()
    title = title_part.split('.')[0] 
    return title

def categorize_age(age):
    if age <= 2:
        return 'baby'
    elif age <= 11:
        return 'kid'
    elif age <= 17:
        return 'adolescent'
    else:
        return 'adult'
```

```python
SeaCareerSet['title'] = SeaCareerSet['name'].apply(extract_title)
SeaCareerSet[['title', 'name']].head()
```

```python
title_distribution = SeaCareerSet['title'].value_counts()
print("Distribution des titres :")
title_distribution
```

```python
SeaCareerSet['group_age'] = SeaCareerSet['age'].apply(categorize_age)

# appliquer la fonctionne et verifier les catégories d'âges
print("Répartition par groupe d'âge :")
print(SeaCareerSet[['age', 'group_age']].head(20))
```

```python
# les passagers mineurs accompagnés
accompanied_minors = SeaCareerSet[(SeaCareerSet['age'] < 18) & ((SeaCareerSet['sibsp'] != 0) | (SeaCareerSet['parch'] != 0))]
print("Passagers mineurs accompagnés:")
accompanied_minors[['name','age','group_age']]
```

```python
print("Type de la colonne avant conversion :",SeaCareerSet['date_embarked'].dtype)
# Convertir la colonne 'date_embarked' en datetime
SeaCareerSet['date_embarked'] = pd.to_datetime(SeaCareerSet['date_embarked'])
# Vérifier le format après conversion
print("Type de la colonne après conversion :",SeaCareerSet['date_embarked'].dtype)
```

```python
# Extraire l'heure d'embarquement
SeaCareerSet['hour_embarked'] = SeaCareerSet['date_embarked'].dt.hour

# Vérifier les heures extraites
print("Heures d'embarquement extraites :")
SeaCareerSet[['date_embarked', 'hour_embarked']].head()
```

```python
# Créer une DataFrame pivot pour analyser les tarifs moyens
pivot_table = SeaCareerSet.pivot_table(values='fare', index=['class', 'sex'], columns='date_embarked', aggfunc='mean')

# Afficher la table pivot
print("Table Pivot des tarifs moyens par classe, sexe et jour d'embarcation :")
pivot_table
```

```python
max_fare_location = pivot_table.stack().idxmax()
print("Classe, sexe et jour d'embarcation avec le tarif moyen le plus élevé :")
print(max_fare_location)
```
